```python
log_2022_df = pd.read_csv(log_2022_path)
log_2022_df.head(2)
```

```python
# 1. UTC 제거
log_2022_df["timestamp_clean"] = log_2022_df["timestamp"].str.replace(" UTC", "", regex=False)

# 2. datetime으로 변환
log_2022_df["timestamp_clean"] = pd.to_datetime(
    log_2022_df["timestamp_clean"],
    format='%Y-%m-%d %H:%M:%S.%f',
    errors='coerce'
)

# 3. NaT가 생긴 경우, 포맷이 다른 문자열을 다시 처리 (마이크로초 없는 경우 등)
mask_failed = log_2022_df["timestamp_clean"].isna()
log_2022_df.loc[mask_failed, "timestamp_clean"] = pd.to_datetime(
    log_2022_df.loc[mask_failed, "timestamp"].str.replace(" UTC", "", regex=False),
    format="%Y-%m-%d %H:%M:%S",
    errors="coerce"
)

# 4. 정리
log_2022_df["timestamp"] = log_2022_df["timestamp_clean"]
log_2022_df.drop(columns="timestamp_clean", inplace=True)
```

```python
log_2023_df = pd.read_csv(log_2023_path)
log_2023_df.head(2)
```

```python
# 1. UTC 제거
log_2023_df["timestamp_clean"] = log_2023_df["timestamp"].str.replace(" UTC", "", regex=False)

# 2. datetime으로 변환
log_2023_df["timestamp_clean"] = pd.to_datetime(
    log_2023_df["timestamp_clean"],
    format='%Y-%m-%d %H:%M:%S.%f',
    errors='coerce'
)

# 3. NaT가 생긴 경우, 포맷이 다른 문자열을 다시 처리 (마이크로초 없는 경우 등)
mask_failed = log_2023_df["timestamp_clean"].isna()
log_2023_df.loc[mask_failed, "timestamp_clean"] = pd.to_datetime(
    log_2023_df.loc[mask_failed, "timestamp"].str.replace(" UTC", "", regex=False),
    format="%Y-%m-%d %H:%M:%S",
    errors="coerce"
)

# 4. 정리
log_2023_df["timestamp"] = log_2023_df["timestamp_clean"]
log_2023_df.drop(columns="timestamp_clean", inplace=True)
```

# 유저 재방문/활성도 분석

## 1. 지원한 유저의 재방문 여부

```python
# 지원일 추출
application_df["application_date"] = pd.to_datetime(application_df["cdate"], format="mixed", errors="coerce")
user_last_application = application_df.groupby("user_uuid")["application_date"].max().reset_index()

# 로그 통합
log_df = pd.concat([log_2022_df, log_2023_df])
log_df["timestamp"] = pd.to_datetime(log_df["timestamp"])

# 유저별 마지막 지원 이후 로그가 있는지 확인
merged = pd.merge(log_df, user_last_application, on="user_uuid", how="inner")
merged["post_application"] = merged["timestamp"] > merged["application_date"]

# 유저별 재방문 여부
user_returned = merged.groupby("user_uuid")["post_application"].max().reset_index()
user_returned.rename(columns={"post_application": "revisited_after_application"}, inplace=True)

# 재방문율
revisit_rate = user_returned["revisited_after_application"].mean() * 100
print(f"지원자 중 재방문한 비율: {revisit_rate:.2f}%")
```

## 2. 지원 이후 로그인 기록이 없는 유저

```python
churn_candidates = user_returned[user_returned["revisited_after_application"] == False]
print(f"지원 이후 플랫폼 재방문 없는 유저 수: {len(churn_candidates)}명")
```

## 3. 북마크만 하고 지원하지 않은 유저 분석

```python
# 북마크 + 지원 여부 확인
bookmark_df = job_bookmark_df.copy()
application_df["application_date"] = pd.to_datetime(application_df["cdate"], errors="coerce", format="mixed")
bookmark_df["bookmark_date"] = pd.to_datetime(bookmark_df["cdate"])

merged = pd.merge(bookmark_df, application_df, on=["user_uuid", "job_uuid"], how="left")
merged["applied"] = merged["application_uuid"].notnull()

non_applied = merged[merged["applied"] == False]["user_uuid"].unique()

# 로그에서 이 유저들의 활동 확인
log_non_applied = log_df[log_df["user_uuid"].isin(non_applied)]
activity_counts = log_non_applied["user_uuid"].value_counts().reset_index()
activity_counts.columns = ["user_uuid", "activity_count"]

# 북마크만 했지만 활동 많은 유저 → 전환 유도 타깃
potential_targets = activity_counts[activity_counts["activity_count"] > 5]
print(f"북마크만 했지만 활동 많은 유저 수: {len(potential_targets)}명")
```
